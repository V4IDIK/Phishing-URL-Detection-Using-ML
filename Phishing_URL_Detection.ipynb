## Split the dataset

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X= df.drop(columns='Result')
X.head()
```

```python
Y=df['Result']
Y=pd.DataFrame(Y)
Y.head()
```

```python
train_X,test_X,train_Y,test_Y=train_test_split(X,Y,test_size=0.30,random_state=10)
```

```python
print("Training set has {} samples.".format(train_X.shape[0]))
print("Testing set has {} samples.".format(test_X.shape[0]))
```

## Desicion Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

#### Model Training

```python
dtree=DecisionTreeClassifier()
model_1=dtree.fit(train_X,train_Y)
```

#### Model Testing

```python
dtree_test=model_1.predict(test_X)
```

```python
print('Accuracy: ', 100.0 * accuracy_score(dtree_test,test_Y))
```

```python
print(classification_report(dtree_test,test_Y))
```

```python
def plot_confusion_matrix(test_Y, predict_y):
 C = confusion_matrix(test_Y, predict_y)
 A =(((C.T)/(C.sum(axis=1))).T)
 B =(C/C.sum(axis=0))
 plt.figure(figsize=(20,4))
 labels = [1,2]
 cmap=sns.light_palette("blue")
 plt.subplot(1, 3, 1)
 sns.heatmap(C, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
 plt.xlabel('Predicted Class')
 plt.ylabel('Original Class')
 plt.title("Confusion matrix")
 plt.subplot(1, 3, 2)
 sns.heatmap(B, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
 plt.xlabel('Predicted Class')
 plt.ylabel('Original Class')
 plt.title("Precision matrix")
 plt.subplot(1, 3, 3)
 sns.heatmap(A, annot=True, cmap=cmap, fmt=".3f", xticklabels=labels, yticklabels=labels)
 plt.xlabel('Predicted Class')
 plt.ylabel('Original Class')
 plt.title("Recall matrix")
 plt.show()
```

```python
plot_confusion_matrix(test_Y, dtree_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate metrics for all models
def calculate_metrics(y_true, y_pred, model_name):
    return {
        'Model': model_name,
        'Accuracy': accuracy_score(y_true, y_pred) * 100,
        'Precision': precision_score(y_true, y_pred, average='weighted') * 100,
        'Recall': recall_score(y_true, y_pred, average='weighted') * 100,
        'F1-Score': f1_score(y_true, y_pred, average='weighted') * 100
    }

# Calculate metrics for each model (replace with your actual predictions)
dtree_metrics = calculate_metrics(test_Y, dtree_test, 'Decision Tree')
rfc_metrics = calculate_metrics(test_Y, rfc_predict, 'Random Forest')
lr_metrics = calculate_metrics(test_Y, logreg_predict, 'Logistic Regression')
knn_metrics = calculate_metrics(test_Y, knn_predict, 'KNN')

# Create comparison dataframe
comparison_df = pd.DataFrame([dtree_metrics, rfc_metrics, lr_metrics, knn_metrics])

print("COMPARISON RESULTS TABLE\n")

print(comparison_df.round(2).to_string(index=False))

print("")

# Plot HeatMap
plt.figure(figsize=(10, 6))
metrics_data = comparison_df.set_index('Model')[['Accuracy', 'Precision', 'Recall', 'F1-Score']]
sns.heatmap(metrics_data, annot=True, fmt='.2f', cmap='YlOrRd', cbar_kws={'label': 'Percentage (%)'})
plt.title('Model Performance Heatmap', fontsize=14, fontweight='bold')
plt.ylabel('Models')
plt.xlabel('Metrics')
plt.show()

# Find the best performing model
best_model = comparison_df.loc[comparison_df['Accuracy'].idxmax()]
print(f"\nBEST PERFORMING MODEL: {best_model['Model']}")

# Performance differences
print(f"\nPerformance Differences:-")
best_accuracy = comparison_df['Accuracy'].max()
for idx, row in comparison_df.iterrows():
    diff = best_accuracy - row['Accuracy']
    if diff > 0:
        print(f"   {row['Model']} is {diff:.2f}% lower than the best model")



```
